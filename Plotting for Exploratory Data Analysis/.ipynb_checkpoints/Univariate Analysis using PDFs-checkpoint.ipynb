<h3>Univariate Analysis - one variable analysis </h3>

- Understand which of these 4 variables (SL, SW, PL, PW) is more useful than other variables to distinguish between Setosa, Versicolor and Virginica.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load iris.csv into pandas DataFrame
iris = pd.read_csv("iris.csv")
```

```python
sns.FacetGrid(iris, hue="species", size = 5)\
    .map(sns.distplot, "petal_length") \
    .add_legend()
    
plt.show();
```

```python
sns.FacetGrid(iris, hue="species", size = 5)\
    .map(sns.distplot, "petal_width") \
    .add_legend()
    
plt.show();
```

- Choice is more for petal_length than petal_width

```python
sns.FacetGrid(iris, hue="species", size = 5)\
    .map(sns.distplot, "sepal_length") \
    .add_legend()
    
plt.show();
```

```python
sns.FacetGrid(iris, hue="species", size = 5)\
    .map(sns.distplot, "sepal_width") \
    .add_legend()
    
plt.show();
```

- The farther the distributions are the better it is.
- The well separated it is the better.

<h4> Conclusion </h4>

- petal_length > petal_width >> sepal_length >> sepal_width
- If I have to pick one feature, I will pick petal_length.
- If I have to pick 2 features, I will pick petal_length & petal_width.
- If I have pick 2 features, I will choose pair plots having features petal_length & petal_width.
