We have a high dimensional space and we want to embed these points in 2-D space such that 
N(x) = {x, x} and N(x) = {x}. 

t-SNE tries to embed these points in low dimensional space such that the distance between the neighboring points is preserved, i.e. 

d(x, x)  d(x', x'), 
d(x, x}  d(x', x'} and 
d(x, x)  d(x', x'). 

But d(x, x)  d(x', x') and d(x, x)  d(x', x')

Neighborhood preserving embedding : t-SNE tries to preserve the distances in the neighborhood. Distances which are outside the neighborhood are not guaranteed to be preserved. 



Mathematical formulation of t-SNE is fairly advanced (2008).
