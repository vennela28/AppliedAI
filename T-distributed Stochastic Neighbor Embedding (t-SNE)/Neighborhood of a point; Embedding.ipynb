<h6> Embedding </h6>

In a d-dimensional space, we want to reduce the dimensions to 2-D or 3-D for visualizing the data. 

Embedding means taking every point one by one and placing them in low dimensional space. 

For every point x in high dimensional space, I am finding point x' in the low dimensional space is called embedding.
