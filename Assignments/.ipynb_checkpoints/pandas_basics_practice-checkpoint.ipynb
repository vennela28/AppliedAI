**OR**

```python
data = {'birds': ['Cranes', 'Cranes', 'plovers', 'spoonbills', 'spoonbills', 'Cranes', 'plovers', 'Cranes', 'spoonbills', 'spoonbills'], \
        'age': [3.5, 4, 1.5, np.nan, 6, 3, 5.5, np.nan, 8, 4], \
        'visits': [2, 4, 3, 4, 3, 4, 2, 2, 3, 2], \
        'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

birds_df = pd.DataFrame(data, index = labels)
```

**4. Print all the rows with only 'birds' and 'age' columns from the dataframe**

```python
birds_df[['birds', 'age']]
```

**5. Select [2, 3, 7] rows and in columns ['birds', 'age', 'visits']**

```python
# Using loc[] - method that takes only index labels and returns rows if the index label exists in the data frame.
# birds_df[['birds', 'age', 'visits']].loc[['c', 'd', 'h']]

# Using iloc[] -  used when the index label of a data frame is something other than numeric series.
birds_df[['birds', 'age', 'visits']].iloc[[2, 3, 7]]
```

**6. Select the rows where the number of visits is less than 4**

```python
birds_df[birds_df['visits'] < 4]
```

**7. Select the rows with columns ['birds', 'visits'] where the age is missing i.e NaN**

```python
# birds_df.isnull() - detects the missing values and returns a boolean value accordingly.

birds_missing = birds_df[birds_df['age'].isnull()]
birds_missing[['birds', 'visits']]
```

**8. Select the rows where the birds is a Cranes and the age is less than 4**

```python
birds_df[(birds_df['birds'] == 'Cranes') & (birds_df['age'] < 4)]
```

**9. Select the rows the age is between 2 and 4(inclusive)**

```python
birds_df[birds_df['age'].between(2,4)]
```

**10. Find the total number of visits of the bird Cranes**

```python
birds_df[(birds_df['birds'] == 'Cranes') & (birds_df['visits'] > 0)].sum()
```

**11. Calculate the mean age for each different birds in dataframe.**

```python
# Using groupby() helps in grouping birds by bird names.
group_birds = birds_df.groupby('birds')

# Using a for-loop to print the mean ages of different category of birds in the DataFrame
for bird, birds_df in group_birds:
    print(bird, birds_df['age'].mean())
```

**12. Append a new row 'k' to dataframe with your choice of values for each column. Then delete that row to return the original DataFrame.**

```python
# Adding a new row with index 'k' to the 'birds' DataFrame
birds_df.loc['k'] = ['kingfisher', 7.0, 5, 'no']
# birds_df

# deleting the newly created row
birds_df = birds_df.drop('k')
birds_df
```

**13. Find the number of each type of birds in dataframe (Counts)**

```python
birds_df['birds'].value_counts()
```

**14. Sort dataframe (birds) first by the values in the 'age' in decending order, then by the value in the 'visits' column in ascending order.**

```python
# Sorting dataframe by 'age' first and then by 'visits'

sorted_df = birds_df.sort_values(['age', 'visits'], ascending = [False, True], axis = 0)
sorted_df

## The below commands sort the dataframe according to the individual column name in the desired order

# birds_df.sort_values(['age'], ascending = False)

# birds_df.sort_values(['visits'], ascending = True)
```
