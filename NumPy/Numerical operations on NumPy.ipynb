# Elementwise Operations

<h3> 1. Basic Operations </h3>

<h4> with scalars </h4>

```python
a = np.array([1, 2, 3, 4]) # creates an array

a + 1
```

```python
a ** 2
```

<h4> All arithmetic operates elementwise </h4>

```python
b = np.ones(4) + 1
b
```

```python
a
```

```python
a - b
```

```python
a * b
```

```python
# Matrix multiplication

c = np.diag([1, 2, 3, 4])

print(c * c)

print('~~~~~~~~~~~~~~~~~~')

print(c.dot(c))
```

<h4> Comparisons </h4>

```python
# element-wise comparisons

a = np.array([1, 2, 3, 4])
b = np.array([5, 2, 2, 4])
a == b
```

```python
a > b
```

```python
# array-wise comparisons

a = np.array([1, 2, 3, 4])
b = np.array([5, 2, 2, 4])
c = np.array([1, 2, 3, 4])

print(np.array_equal(a, b))
print(np.array_equal(a, c))
```

<h4> Logical Operations </h4>

```python
a = np.array([1, 1, 0, 0], dtype=bool)
b = np.array([1, 0, 1, 0], dtype=bool)

np.logical_or(a, b)
```

```python
np.logical_and(a, b)
```

Data in populations.txt describes the populations of hares and lynxes (and carrots) in northern Canada during 20 years.

```python
# load data into numpy array object
data = np.loadtxt('populations.txt')
```

```python
data
```

```python
year, hares, lynxes, carrots = data.T   # columns to variables
print(year)
```

```python
# The mean population over time
populations = data[:, 1:]  # : - take all the rows, 1: - take from 1st column
populations
```

```python
# sample standard deviations
populations.std(axis=0)
```

```python
# which species has the highest population each year?
np.argmax(populations, axis = 1)   # axis = 1 - operate row-wise
```

<h3> Array Shape Manipulation </h3>

<h4> Flattening </h4>

```python
a = np.array([[1, 2, 3], [4, 5, 6]])
a.ravel()  # Returns a contiguous flattened array. 
```

```python
a.T  # transpose
```

```python
a.T.ravel()
```

<h4> Reshaping </h4>

The inverse operation to flattening.

```python
print(a.shape)
print(a)
```

```python
b = a.ravel()
print(b)
```

```python
b = b.reshape((2, 3))
b
```

```python
b[0, 0] = 100
a
```

<h4> reshape may also return a copy! </h4>

```python
a = np.zeros((3, 2))
b = a.T.reshape(3*2)
b[0] = 50
print(a)
print(b)
```
