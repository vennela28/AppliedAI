# Introduction to Pandas dataframe

Data frame is a main object in Pandas. It is used to represent data with rows and columns.

Data frame is a data structure representing the data in tabular / excel spread sheet.

<h3> Creating Dataframe </h3>

```python
import pandas as pd
df = pd.read_csv('weather_data.csv')   # read csv data
df
```

```python
# list of tuples

weather_data = [('1/1/2017', 32, 6, 'Rain'),
                ('1/2/2017', 35, 7, 'Sunny'),
                ('1/3/2017', 28, 2, 'Snow'),
                ('1/4/2017', 24, 7, 'Snow'),
                ('1/5/2017', 32, 4, 'Rain'),
                ('1/5/2017', 31, 2, 'Sunny')
               ]

df = pd.DataFrame(weather_data, columns=['day', 'temperature', 'windspeed', 'event'])
df
```

```python
# get dimensions of data

df.shape   # total number of rows and columns
```

```python
# if we want to see the intial few rows use head()  -  default = 5 rows
df.head()
```

```python
# if we want to see the last few rows use tail() - default = 5 rows
df.tail()
```

```python
# slicing
df[2:5]
```

```python
# print columns in a table
df.columns
```

```python
# print particular column data - 1st way
df.day
```

```python
# print particular column data - 2nd way
df['day']
```

```python
# get data from 2 columns
df[['day', 'event']]
```

```python
# get all temperatures
df['temperature']
```

```python
# get minimum temperature
df['temperature'].min()
```

```python
# get maximum temperature
df['temperature'].max()
```

```python
# describe temperature data
df['temperature'].describe()  # we get 6 values - mean, std. deviation, statistical parameters(25%, 50%, 75%), min, max
```

```python
# select rows which has maximum temperature
df[df.temperature == df.temperature.max()]
```

```python
# select only the day column which has maximum temperature
df.day[df.temperature == df.temperature.max()]
```
